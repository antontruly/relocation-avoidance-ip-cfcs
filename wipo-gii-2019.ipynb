```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
sns.set(rc={"figure.dpi": 300, "savefig.dpi": 300})
sns.set_context("notebook")
```

```python
sns.set(rc={"figure.figsize": (8, 5)})
sns.set_style("ticks")
plt.rcParams["font.family"] = "Times New Roman"
```

```python
wipo_gii_tax_2019 = pd.read_csv(
    "wipo_gii_tax_2019.csv",
    index_col=False,
)
```

```python
wipo_gii_tax_2019_trunc = wipo_gii_tax_2019.copy()
wipo_gii_tax_2019_trunc.loc[(
    (wipo_gii_tax_2019_trunc.innovation_input>44)
    & (wipo_gii_tax_2019_trunc.innovation_output>24)
    & (wipo_gii_tax_2019_trunc.innovation_input<58)
    & (wipo_gii_tax_2019_trunc.innovation_output<45)), "iso_2"
] = ""
```

```python
ax = sns.scatterplot(
    x="innovation_input",
    y="innovation_output",
    hue="tax",
    size="tax",
    alpha=.8,
    sizes=(10, 100),
    data=wipo_gii_tax_2019_trunc
)
for i in range(0, wipo_gii_tax_2019_trunc.shape[0]):
    ax.text(
        wipo_gii_tax_2019_trunc.innovation_input[i]+.5,
        wipo_gii_tax_2019_trunc.innovation_output[i]+.3,
        wipo_gii_tax_2019_trunc.iso_2[i],
        horizontalalignment="left",
        size="small",
        color="black",
    )
ax.legend(title='Tax',loc='upper left')
ax.set(xlabel='Innovation Input', ylabel='Innovation Output')
```

```python
sns.set(rc={"figure.figsize": (6, 4.5)})
sns.set_style("ticks")
plt.rcParams["font.family"] = "Times New Roman"
```

```python
ax_ = sns.scatterplot(
    x="innovation_input",
    y="innovation_output",
    hue="tax",
    size="tax",
    alpha=0.8,
    legend=False,
    sizes=(10, 100),
    data=wipo_gii_tax_2019,
)
for line in range(0, wipo_gii_tax_2019.shape[0]):
    ax_.text(
        wipo_gii_tax_2019.innovation_input[line] + 0.1,
        wipo_gii_tax_2019.innovation_output[line] - 0.1,
        wipo_gii_tax_2019.iso_2[line],
        horizontalalignment="left",
        size="small",
        color="black",
    )
ax_.set(
    xlabel="Innovation Input", ylabel="Innovation Output", ylim=(24, 45), xlim=(44, 58)
)
```
